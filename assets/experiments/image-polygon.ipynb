```python
from skimage.draw import polygon
import numpy as np
img = np.zeros((10,10),dtype=np.uint8)
r = np.array([1,2,8])
c = np.array([1,7,4])
rr,cc = polygon(r,c)
img[rr,cc]=3
```

```python
from skimage.draw import polygon_perimeter
rr,cc = polygon_perimeter(r,c,shape=img.shape,clip=True)
img[rr,cc]=2
img[r,c]=1
```

```python
im = Image.fromarray(img)
plt.imshow(im)
```
